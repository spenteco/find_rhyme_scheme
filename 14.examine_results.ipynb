```python
import pandas as pd

metadata = pd.read_csv('earlyprint_metadata.csv')

metadata
```

```python
report_lines = open('OUT.14.txt').read().split('\n')
```

```python
import re

results_by_tcp_id = {}

for line in report_lines:
    
    if line.startswith('OK ') or line.startswith('ERROR '):
        
        line_parts = [t for t in re.split('\s+', line) if t > '']
        
        ok_or_error = line_parts[0]
        tcp_id = line_parts[1]
        rhyme_scheme = line_parts[3]
        
        if tcp_id not in results_by_tcp_id:
            results_by_tcp_id[tcp_id] = []
        results_by_tcp_id[tcp_id].append(rhyme_scheme)
        
        all_rhyme_schemes.append(rhyme_scheme)
            
print(len(results_by_tcp_id))
```

```python
from collections import Counter
    
tcp_id_rhyme_scheme = []
all_rhyme_schemes = []

for tcp_id, rhyme_schemes in results_by_tcp_id.items():
    
    if len(rhyme_schemes) > 49:
        
        all_rhyme_schemes += rhyme_schemes
    
        m = metadata.loc[metadata['TCP ID'] == tcp_id].to_dict('records')
    
        if len(m) > 0:

            for w in Counter(rhyme_schemes).most_common():

                tcp_id_rhyme_scheme.append({'tcp_id': tcp_id, 
                                       'date': m[0]['Date'], 
                                       'author': m[0]['Author'], 
                                       'title': m[0]['Title'],
                                       'rhyme_scheme': w[0],
                                       'n_stanzas': w[1],
                                       'rel_freq_stanzas': (float(w[1]) / float(len(rhyme_schemes)))})
                
rhyme_scheme_counts_output = []
for w in Counter(all_rhyme_schemes).most_common():
    rhyme_scheme_counts_output.append({'rhyme_scheme': w[0], 'n': w[1]})
    
print('ok')
```

```python
df = pd.DataFrame(tcp_id_rhyme_scheme)

column_names = ['tcp_id', 'date', 'rhyme_scheme', 
                'n_stanzas', 'rel_freq_stanzas', 
                'date', 'author', 'title']

df = df.reindex(columns=column_names)

df.to_csv('tcp_id_rhyme_scheme.14.csv', index=False)

#

df = pd.DataFrame(rhyme_scheme_counts_output)

#column_names = ['tcp_id', 'date', 'rhyme_scheme', 
#                'n_stanzas', 'rel_freq_stanzas', 
#                'date', 'author', 'title']

#df = df.reindex(columns=column_names)

df.to_csv('rhyme_scheme_counts.14.csv', index=False)


print('ok')
```
