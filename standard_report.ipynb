```python
CONTROLS = {'n_lines': 7, 'standard_rhyme_schemes': ['ababbcc',]}
#CONTROLS = {'n_lines': 9, 'standard_rhyme_schemes': ['ababbcbcc',]}
#CONTROLS = {'n_lines': 14, 'standard_rhyme_schemes': ['abbaabbacdecde', 'abbaabbacdcdcd', 'ababcdcdefefgg']}
```

```python
import re
import pandas as pd

metadata = pd.read_csv('earlyprint_metadata.csv')

by_rhyme_scheme = {}

for line in open('OUT.' + str(CONTROLS['n_lines']) + '.txt').read().split('\n'):
    if line.startswith('OK ') or line.startswith('ERROR '):
        
        line_parts = [p for p in re.split('\s+', line.strip()) if p > '']
        
        tcp_id = line_parts[1]
        stanza_n = line_parts[2]
        rhyme_scheme = line_parts[3]
        terminal_words = eval(line[line.find(' ['):].strip().replace('] [', ']] [[').split('] [')[0])
        
        if rhyme_scheme not in by_rhyme_scheme:
            by_rhyme_scheme[rhyme_scheme] = []
    
        m = metadata.loc[metadata['TCP ID'] == tcp_id].to_dict('records')
            
        by_rhyme_scheme[rhyme_scheme].append({'tcp_id': tcp_id, 
                                                'date': m[0]['Date'], 
                                                'author': m[0]['Author'], 
                                                'title': m[0]['Title'],
                                                'stanza_n': stanza_n, 
                                                'terminal_words': terminal_words})
        
for rhyme_scheme, stanzas in sorted(by_rhyme_scheme.items(), key=lambda item: len(item[1]), reverse=True):

    flag = ''
    if rhyme_scheme in CONTROLS['standard_rhyme_schemes']:
        flag = 'STANDARD'
    else: 
        unique_rhyme_letters = ''.join(sorted(list(set(rhyme_scheme))))

        if len(unique_rhyme_letters) == len(rhyme_scheme):
            flag = 'BLANK'
            
        elif len(rhyme_scheme) % 2 == 0 and rhyme_scheme in 'aabbccddeeffgghhii':
            flag = 'COUPLETS'

    print(rhyme_scheme, len(stanzas), len(unique_rhyme_letters), flag)
```
